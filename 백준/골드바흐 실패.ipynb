```python
def isPrime(num):
    if num == 1:
        return False
    else:
        for i in range(2,int(num**0.5)+1):
            if num%i == 0:
                return False
        return True
```

```python
def isPrime_list(N):
    isPrime_list = []
    for i in range(2,N):
        if isPrime(i):
            isPrime_list.append(i)
    
    return isPrime_list
            
    
```
